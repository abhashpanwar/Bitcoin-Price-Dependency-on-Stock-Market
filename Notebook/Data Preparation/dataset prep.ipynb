```python
import pandas as pd
import numpy as np
import datetime
```

```python
df_main = pd.read_csv("../../data/bitcoin gold price.csv")
df_main.head()
```

```python
df_main['Date'] = pd.to_datetime(df_main['Date'])
```

```python
df_vix = pd.read_csv('../../data/vixcurrent.csv', header=1)
df_vix.head()
```

```python
df_vix['Date'] = pd.to_datetime(df_vix['Date'])
df_vix.head()
```

```python
df_vix = df_vix.drop(['VIX Open', 'VIX High', 'VIX Low'], axis=1)
```

```python
df_vix.head()
```

```python
df_final = pd.merge(df_main, df_vix, how='left', on='Date')
```

```python
df_final.head(5)
```

```python
df_final.describe()
```

```python
df_final.isnull().sum()
```

```python
df_final.dropna(inplace=True)
df_final.reset_index(drop=True, inplace=True)
```

```python
df_final.head(10)
```

```python
df_final = df_final.drop(['M2(Not seasonally adjusted)[1e+09 USD]'], axis=1)
```

```python
df_final.to_csv('../../data/final_dataset.csv')
```
